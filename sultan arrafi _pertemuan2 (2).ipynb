```python
# prompt: getsum getkali list =1,3,5,7

def getsum(list):
  total = 0
  for i in list:
    total += i
  return total

def getkali(list):
  total = 1
  for i in list:
    total *= i
  return total

list = [1, 3, 5, 7]
print("Sum:", getsum(list))
print("Multiplication:", getkali(list))
```
